Splitting the Features and Target

```python
X=heart_data.drop(columns='target',axis=1)
Y=heart_data['target']
```

```python
print(X)
```

```python
print(Y)
```

splitting the data into training data and test data

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model Training

```python
model=LogisticRegression()
```

```python
#training the logisticsRegression Model
```

```python
model.fit(X_train,Y_train)
warnings.filterwarnings("ignore", category=UserWarning)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler().fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
mean_std_values = {'mean': scaler.mean_, 'std': scaler.scale_}
with open('mean_std_values.pkl', 'wb') as f:
    pickle.dump(mean_std_values, f)
```

Model Evalution

<!-- Accuracy Score -->
print(X_train)

```python
print(X_train)
```

```python
X_train_prediction=model.predict(X_train)
training_data_accuracy=  accuracy_score(X_train_prediction,Y_train)
warnings.filterwarnings("ignore", category=UserWarning)
```

```python
print('Accuracy on Training data:', training_data_accuracy)
```

```python
X_test_prediction=model.predict(X_test)
test_data_accuracy=  accuracy_score(X_test_prediction,Y_test)
warnings.filterwarnings("ignore", category=UserWarning)

```

```python
print('Accuracy on Test data:', test_data_accuracy)
```

Building the predicting the system

```python
input_data=(48,1,0,122,222,0,0,180,0,0,2,0,2)
#change the input data to a numpy array
```

```python
input_data_as_numpy_array=np.asarray(input_data)
print(input_data_as_numpy_array)
warnings.filterwarnings("ignore", category=UserWarning)
input_data_reshaped=input_data_as_numpy_array.reshape(1,-1)
prediction=model.predict(input_data_reshaped)
print(prediction)
if(prediction[0]==0):
    print('The person does not  a heart Disease')
else:
    print('The person has heart Disease')
```
